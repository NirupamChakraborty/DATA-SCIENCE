```python
### Accessing List Elements

fruits=["apple","banana","cherry","kiwi","gauva"]

print(fruits[0])
print(fruits[2])
print(fruits[4])
print(fruits[-1])
```

```python
print(fruits[1:])  #this means print from 1 to end
print(fruits[1:3]) # print from 1 to 2 and exclude 3 index element
# these slicing methods will return a list
print(fruits[-1:])
print(fruits[-1:-3]) # [] confusing indexing
print(fruits[:-1])
```
