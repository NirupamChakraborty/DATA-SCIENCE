```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
### Variable Length Arguments
## Positional And Keywords arguments

def print_numbers(*krish):
    for number in krish:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Krish")
```

```python

## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Krish")
```

```python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Krish",age="32",country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Krish",name="Krish",age="32",country="India")
```

```python
def convert_temp(temp,unit):
    if unit=="C":
        return (temp*9/5)+32
    elif unit=="F":
        return (temp-32)*5/9
    else:
        return "Invalid unit"  

convert_temp(25,"C")              
```

```python
# Password checker
def is_strong_password(password):
    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False 
    if not any(char.isupper() for char in password):
        return False
    if not any(char in "!@#$%^&*()-_=+[]{}|;:'\",.<>?/`~" for char in password):
        return False 
    return True


print(is_strong_password("Krish@123"))
print(is_strong_password("123456789@Pass"))
```

```python
# palindrome
def is_pal(s):
    s=s.lower().replace(" ","")
    return s[::-1]==s
print(is_pal("Madam"))
```

```python
# palindrome
def is_pal(s):
    rev =""
    end =len(s)-1
    
    while end >= 0:
        rev = rev+s[end]
        end-=1
    return rev.lower() == s.lower()  
    
print(is_pal("Madam"))
```
